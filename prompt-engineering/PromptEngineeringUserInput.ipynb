```python
from google import genai
import json
# The client gets the API key from the environment variable `GEMINI_API_KEY`.

# Step 1: Get input from the user
name = input("Enter your name: ")
college = input("Enter your college: ")
experience = input("Enter your years of experience: ")
project = input("Enter your current project: ")
location = input("Enter your location: ")

# Step 2: Construct the prompt
user_info = {"name": name,
             "college": college,
             "experience": experience,
             "project": project,
             ""
            }

client = genai.Client(api_key="YOUR_GEMINI_API_KEY")
prompt = f"You will receive input values. Your task is to print them as a JSON object. Return only the JSON response and no additional explanation or text Json data is {json.dumps(user_info)}"

# This will also work
# prompt = f"You will receive input values. Your task is to print them as a JSON object. Return only the JSON response and no additional explanation or text Json data is {user_info}"

response = client.models.generate_content(
    model="gemini-2.5-flash", contents=prompt
)

```

```python
print(response.text)
```
