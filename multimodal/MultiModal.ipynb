``````````````python
import ollama

query = ''
message = [
    {  # Attach the image only once at the start
        'role': 'user',
        'content': 'How many people in this image?',
        'images': ['street.jpg']
    }
]

# Initial request to process the image
response = ollama.chat(
    model='llama3.2-vision',
    messages=message
)
message.append({
    'role': 'assistant',
    'content': response['message']['content']
})
print(response['message']['content'])

# Loop for further queries — no need to attach the image again
while query != 'exit':
    query = input('Enter your query: ')
    if query == 'exit':
        break

    # Append user query without adding the image again
    message.append({
        'role': 'user',
        'content': query
    })

    # Send chat request
    response = ollama.chat(
        model='llama3.2-vision',
        messages=message
    )

    # Append assistant response
    message.append({
        'role': 'assistant',
        'content': response['message']['content']
    })

    # Print assistant's response
    print(response['message']['content'])
``````````````

### Multiple images

``````````````python
from google import genai
from google.genai import types

client = genai.Client(api_key="YOUR_GOOGLE_API_KEY")

# Upload the first image
image1_path = "s1.png"
uploaded_file = client.files.upload(file=image1_path)

# Prepare the second image as inline data
image2_path = "s2.png"
with open(image2_path, 'rb') as f:
    img2_bytes = f.read()

# Create the prompt with text and multiple images
response = client.models.generate_content(

    model="gemini-2.5-flash",
    contents=[
        "What is different between these two images?",
        uploaded_file,  # Use the uploaded file reference
        types.Part.from_bytes(
            data=img2_bytes,
            mime_type='image/png'
        )
    ]
)

print(response.text)
``````````````

### Text to Image

``````````````python
from google import genai
from google.genai import types
from PIL import Image
from io import BytesIO
import base64

client = genai.Client(api_key="YOUR_GOOGLE_API_KEY")

contents = ('Hi, Create an Image of polar bears playing guitar , drums , piano and one on vocals')

response = client.models.generate_content(
    model="gemini-2.0-flash-preview-image-generation",
    contents=contents,
    config=types.GenerateContentConfig(
      response_modalities=['TEXT', 'IMAGE']
    )
)

for part in response.candidates[0].content.parts:
  if part.text is not None:
    print(part.text)
  elif part.inline_data is not None:
    image = Image.open(BytesIO((part.inline_data.data)))
    image.save('gemini-native-image.png')
    image.show()
``````````````

### text-image to image

``````````````python
from google import genai
from google.genai import types
from PIL import Image
from io import BytesIO

import PIL.Image

image = PIL.Image.open('castle.jpeg')

client = genai.Client(api_key="YOUR_GOOGLE_API_KEY")

text_input = ('Hi, This is a picture ofa castle'
            'Can you make it to winter and add christmas tree decorated.'
              'DO NOT TRANSFORM ANYTHING , JUST CHANGES IN ORGINAL PHOTO')

response = client.models.generate_content(
    model="gemini-2.0-flash-preview-image-generation",
    contents=[text_input, image],
    config=types.GenerateContentConfig(
      response_modalities=['TEXT', 'IMAGE']
    )
)

for part in response.candidates[0].content.parts:
  if part.text is not None:
    print(part.text)
  elif part.inline_data is not None:
    image = Image.open(BytesIO((part.inline_data.data)))
    image.show()
``````````````

### Document Understanding

``````````````python
from google import genai
from google.genai import types
import pathlib

client = genai.Client(api_key="YOUR_GOOGLE_API_KEY")

# Retrieve and encode the PDF byte
filepath = pathlib.Path('Telecom_Fault_Management_Policy.pdf')

prompt = "get me the Fault escalation manager CONTACT"
response = client.models.generate_content(
  model="gemini-2.5-flash",
  contents=[
      types.Part.from_bytes(
        data=filepath.read_bytes(),
        mime_type='application/pdf',
      ),
      prompt])
print(response.text)
``````````````
