```python
from google import genai

# The client gets the API key from the environment variable `GEMINI_API_KEY`.
client = genai.Client(api_key="YOUR_GEMINI_API_KEY")
```

```python
descriptionText = """Sarah works in an office as a project coordinator. She usually arrives at work by 9:00 AM. The first thing she does is check her emails. She replies to important messages and makes a to-do list for the day.

At 10:00 AM, she attends a daily team meeting. In the meeting, her manager gives updates on ongoing projects. Sarah takes notes and asks questions if something is unclear.

After the meeting, Sarah works on preparing a report for a client. She collects data from her teammates and puts everything into a PowerPoint presentation. She also double-checks the numbers to make sure they are correct.

Around 1:00 PM, Sarah takes a lunch break with her coworkers. After lunch, she continues working on her tasks. Sometimes, she has video calls with clients in the afternoon.

Before leaving at 5:30 PM, Sarah reviews her to-do list and checks off the things she has finished. She updates her manager and sets reminders for anything she needs to do the next day.

"""
```

```python

query = '' #initlaise query
messages = [descriptionText]
while  query!='exit': 
    query = input("Ask me question?")
    if query.lower() == 'exit':
        print("Chatbot: Goodbye!")
        break
    messages.append(query)
    response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents=messages,
    )
    print(response.text)
```
