### Fuzzy Input

```python
system_message = f'''You are a helpful parsing assistant. You will be supplied with an image containing personal details.

Your task is to carefully extract and parse the following details from the image:
- Name
- House number
- Email address
- Phone number

You must return the extracted information strictly in the following JSON format:

{{
    "name": "john-wick",
    "house_number": "123-green park",
    "email_id": "john-wick@example.com",
    "phone_number": "+91-98765433"
}}

Only provide the response in JSON format — nothing else.

Always end the response with: "Good day."
'''
```

```python
query='provide the details'
```

```python
response = ollama.chat(model='llama3.2-vision',
        messages=[{'role':'system','content':system_message},
                {'role':'user','content':query,'images':['pic1.png']}]
        
       )
```

```python
print(response.message.content)
```

```python
response = ollama.chat(model='llama3.2-vision',
        messages=[{'role':'system','content':system_message},
                {'role':'user','content':query,'images':['pic2.png']}]
        
       )
```

```python
print(response.message.content)
```

```python
import ollama
```

```python
response = ollama.chat(
    model='llama3.2',
    messages=[{'role':'user','content':query}]
)

```

```python
response.message.content
```

```python
query = '' #initlaise query
messages=[{'role':'system','content':'you are a freindly chatbot but always limit your response within 15 words'}] #memory



while  query!='exit': #run untill exit
    query=input('Enter your query: ' )
    messages.append({'role':'user','content':query})
    response=ollama.chat(
        model='llama3.2',
        messages=messages)
    result = response.message.content
    messages.append({'role':'assistant','content':result})
    print(result)
    print('------------------')
```

```python
query = ''

path='pic1.png'

messages=[{'role':'system','content':'you are a careful observer','images':[path]}]


while  query!='exit':
    query=input()
    messages.append({'role':'user','content':query})
    response=ollama.chat(
        model='llama3.2-vision',
        messages=messages) #messages or memory
    result = response.message.content
    messages.append({'role':'assistant','content':result})
    print(result)
```
