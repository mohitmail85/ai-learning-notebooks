# Embeddings

- Vector search is a common way to store and search over unstructured data (such as unstructured text).
- The idea is to **store numeric vectors that are associated with the text**.
- **Given a query, we can embed it as a vector of the same dimension and use vector similarity metrics (such as cosine similarity) to identify related text**.

```python
from langchain_huggingface import HuggingFaceEmbeddings

embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-mpnet-base-v2")
```

# Vectore Stores

* Vector store is a common way to store and search unstructured data. The unstructred data is converted to Numeric vectors or embeddings and then store. So whenever we want to retreive a data we can easily  **EMBEDED** our input into vector and through similairty search we can retreive answer.

* Vector stores are initialised with embedding models such as how text data is stored in DB

* Note that most vector store implementations will allow you to connect to an existing vector store-- e.g., by providing a client, index name, or other information. See the documentation for a specific integration for more detail

```python
from langchain_chroma import Chroma
vectorestore = Chroma.from_documents(
                    documents,
        embedding=embeddings,
    persist_directory = 'sunday/chorma/'
)
```

### Documents are instantialted
 we can quesry 
 > * Synchronously or asynchronously
 > * By string query and by vector
 > * With and without returning similarity scores;
 > * By similarity and maximum marginal relevance (to balance similarity with query to diversity in retrieved results)

```python
#BAASED on similarity search

vectorestore.similarity_search('cat',k=3)
```

```python
#BAASED on similarity search

vectorestore.similarity_search('game',k=3)
```

# Retreivers

###  Runnables

```python
from langchain_core.runnables import RunnableLambda

retreiver = RunnableLambda(vectorestore.similarity_search).bind(k=1)


retreiver.invoke('cat')
```

```python
retreiver.batch(['cat','ice','snow'])
```

```python
retreiver.batch(['snow'])
```

LangChain VectorStore objects do not subclass Runnable, and so **cannot immediately be integrated into LangChain Expression Language chains.**

LangChain Retrievers are Runnables, so they implement a standard set of methods (e.g., synchronous and asynchronous invoke and batch operations) and **are designed to be incorporated in LCEL chains.**

### Through vectorestores as retreivers

```python
vectorestore.as_retriever(search_type="similarity",
    search_kwargs={"k": 1}
                 ).invoke('monster')
```
