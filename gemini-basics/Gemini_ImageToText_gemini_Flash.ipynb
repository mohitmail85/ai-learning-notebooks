```python
from google import genai

# The client gets the API key from the environment variable `GEMINI_API_KEY`.
client = genai.Client(api_key="YOUR_GEMINI_API_KEY")
```

```python
my_file = client.files.upload(file="idcard.png")
```

```python
prompt = """You are image parser, your job is to parse the image and provide the name, date of birth and id no details of the image in json format. 
Don't print anything other than json
"""
response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents=[my_file, prompt ],
)
```

```python
print(response.text)
```
